```python
import pandas as pd
import numpy as np

#--- การจัดกลุ่มสำหรับ Object Series ---
data = [5,15,10,15,20,15,10,5]
index = ['a','b','a','c','c','d','b','c']

sr = pd.Series(data, index=index)
print(
    sr.sort_index(),#แแสดงข้อมูลโดยให้เรียงตาม Index 
    sr.groupby(index).sum(), #แสดงผล Sum ตาม Index 
    sr.groupby(index).count(), #นับจำนวนข้อมูลของ Index
    sr.groupby(index).mean(), #หาค่าเฉลี่ย
    sr.groupby(index).min(), #หาค่าน้อยสุด
    sr.groupby(index).max() #หาค่ามากสุด
) 
```

```python
# -- การจัดกลุม DataFrame โดยใช้ Column เดียว --
data2 = {
    'Group' : ['B','C','B','A','A','C','B','C','A'],
    'Name' : ['Jim', 'Aon', 'Ya', 'Aey', 'Bouy', 'P_Som', 'Tou', 'Park', 'Nuy'],
    'Score' : [70,60,80,50,70,50,90,70,80]
}

df = pd.DataFrame(data2)
print(
    df.sort_values('Group'),
    df.groupby('Group').sum(),
    df.groupby('Group').count(),
    df.groupby('Group').mean(),
    df.groupby('Group').max(),
    df.groupby('Score').max(),
    df.groupby('Group').min()
)
```
