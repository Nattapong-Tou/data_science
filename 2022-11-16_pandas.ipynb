```python
# *** Pandas ***
# import libraries
import pandas as pd
import numpy as np
```

```python
# ทดสอบ plot
sr = pd.Series([10,20,30,50,20,40,20])
sr.plot.line()
```

```python
# ทดสอบ plot grap with properties
data = np.random.randint(10,101,10)
sr_2 = pd.Series(data, index=list('abcdefghij'))

sr_2.plot(kind='line', title='Line Plot With Pandas',style='ro--', grid=True,fontsize=16)
```

```python
# ทดสอบกำหนดค่าแบบกำหนดค่า แกน x,y
data3 = np.random.randint(10,101,10)
sr3 = pd.Series(data, index=list('abcdefghij'))

sr3.plot(kind='line',style='mD:', grid=True, use_index=False,
    xticks=np.arange(0, 10, 1),
    yticks=np.arange(10, 110, 10)
)
```

```python
# เลือกเฉพาะจำนวนที่มีค่าตั้งแต่ 50 ขึ้นไปมา Plot
data4 = np.random.randint(10,101,10)
sr4 = pd.Series(data4)

sr4[sr4 > 50].plot(kind='line',style='bp--', grid=True, use_index=False)
```

```python
# เลือกเฉพาะ 10 รายการสุดท้ายของข้อมูลมา Plot
data5= np.random.randint(10,101,10)
sr5 = pd.Series(data4)

sr5.tail(10).plot(kind='line',style='bp--', grid=True, use_index=False)
```

```python
# *** การแสดงกราฟข้อมูลหลายชุด ***
y_tk = np.arange(1.0, 11, 1.5) # tick ในแนวแกน y 10 จำนวน ระยะห่าง 1.5

data6 = [3.75, 4.8, 9.7, 1.50, 6.3, 4.4, 5.5]
sr6 = pd.Series(data6)
sr6.plot(kind='line', style='bo--', grid=True, yticks=y_tk, legend=True, label='Data1')

data7 = np.random.randint(1, 11, 7)
sr7 = pd.Series(data7)
sr7.plot(kind='line', style='mD:', alpha=0.6, grid=True, yticks=y_tk, legend=True, label='Data2')
```
