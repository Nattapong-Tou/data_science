```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # import module สำหรับสร้างกราฟ 3D

```

```python
# Bar Chart
#การวาดกราฟแท่ง
y = [5,15,15,20,10]
x = ['Mon','Tue','Wed','Thu','Fri']

plt.bar(x, y, color='y', alpha=0.8)
plt.xticks(rotation=45)
plt.show()
```

```python
#การวาดกราฟหลายเส้น
x = [5,15,15,20,10]
y = [1,2,3,4,5]

plt.plot(y, x, ls='-', c='r', label='Product #1')

x = [15,20,10,5,15] # เปลี่ยนค่าแกน x ใหม่ ส่วนแกน y คงเดิม
plt.plot(y, x, ls='--', c='g', label='Product #2')

x = [10,5,15,15,15] # เปลี่ยนค่าแกน x ใหม่ ส่วนแกน y คงเดิม
plt.plot(y, x, ls=':', c='b', label='Product #3')

plt.style.use('seaborn')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.style.use("ggplot")
y = [2,4,1,2,3,2,1,2,3,2]
x = np.arange(1,11)

plt.bar(x, y, color='m')
plt.title("Bar Plot")
plt.xlabel("Play")
plt.ylabel("Score")
plt.grid(color='y', alpha=0.2)
plt.show()
```

```python
# วาดกราฟ Bar Plot แบบข้อมูลหลายชุด

y = [10, 15, 15, 20, 10]
x = np.arange(1,6)

bw = 0.25 #ความกว้างของกราฟแต่ล่พแท่ง

# วาดกราฟแท่งแรกตามปกติ
plt.bar(x, y, color='g', width=bw, label='Product#1')

# กราฟอันที่สอง เลื่อนตำแหน่งในแนวแกน x เท่ากับความกว้างของแท่งกราฟ (อันแรก)
y = [15,25,10,15,20]
plt.bar(x+bw, y, color='lightblue',width=bw, label='Product#2')

# กราฟอันที่สาม เลื่อนตำแหน่งแนวแกน x เท่ากับความกว้างของแท่งกราฟ (อันแรกกับอันที่สองรวมกัน)
y = [10,20,15,10,15]
plt.bar(x+(2*bw), y, hatch='..', color='lightgray', width=bw, label='Product#3')

# เลื่อนระยะ tick ในแนวแกน x เพื่ออยู่ศูนย์กลางของกราฟ
plt.xticks(x+bw, [2016, 2017, 2018, 2019, 2020])

plt.style.use('seaborn')
plt.legend('best')
plt.show()
```

```python
# Pie Chart
data = [50,20,15,5,10]
lb = ['Chrome', 'Edge', 'Firefox', 'Opera', 'Safari']

plt.pie(data, labels=lb, colors=['#40ff00', '#0066ff', '#ff6600', '#cc0000', '#99ccff'])
plt.show()
```

```python
data = [50,20,15,5,10]
lb = ['Chrome', 'Edge', 'Firefox', 'Opera', 'Safari']
explode = [0, 0.1, 0, 0.2, 0] # ระยะแยกจากกราฟหลัก

plt.pie(data, labels=lb, explode=explode, startangle=30,
    colors=['#40ff00', '#0066ff', '#ff6600', '#cc0000', '#99ccff'], shadow=True, autopct="%.1f%%")
plt.show()
```

```python
# การวาดกราฟแบบ Histogram
plt.style.use('seaborn')

data = [22,24,24,25,25,21,23,24,23,25,25,22,20,22,
    28,27,29,29,30,26,27,29,30,30,29,29,
    33,34,34,32,31,31,34,35,33,32,32,34,
    37,39,36,38,38,37,40,39,40,37,
    45,44,43,43,41,42,44,46,43,44,
    50,47,48,47,46,49,50,48,46,
    52,53,55,51,51,54,55,52,
    59,60,58,58,56]

plt.hist(data, bins=8, color='m', edgecolor='k', alpha=0.5)
plt.show()
```

```python
# การวาดกราฟแบบ Scatter
plt.style.use('seaborn')

y = np.random.randint(1, 101, 20)
x = np.arange(1, 21)

plt.scatter(x, y, color='g', marker='o')
plt.yticks(np.arange(0, 101, 10))
plt.xticks(x)

plt.show()
```

```python
# การวาดกราฟแบบ Scatter
plt.style.use('seaborn')

salary = [15000, 17000, 22000, 23000, 32000,
    45000, 24000, 30000, 35000, 38000]

gender = ['Male', 'Male', 'Female', 'Female', 'Female',
    'Male', 'Male', 'Female', 'Male', 'Female']

plt.scatter(gender, salary, color='m', marker='D', s=100)
plt.yticks(np.arange(10000, 50000, 5000))

plt.show()
```

```python
# การวาดกราฟแบบ 3D

ax = plt.axes(projection='3d')

x = np.random.randint(10, 20, 5)
y = np.random.randint(30, 40, 5)
z = np.random.randint(1, 5, 5)

ax.plot3D(x,y,z, 'ro--')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
