```python
#การบันทึกข้อมูลหลายข้อมูลลงในแต่ล่ะ Sheet
data1 = {
    'ONE' : np.random.randint(10, 100, 5),
    'TWO' : np.random.randint(10, 100, 5),
    'THREE' : np.random.randint(10, 100, 5)
}
df6 = pd.DataFrame(data1)

data2 = {
    '1' : np.random.randint(1, 100, 5),
    '2' : np.random.randint(1, 100, 5),
    '3' : np.random.randint(1, 100, 5)
}
df7 = pd.DataFrame(data2)

# เขียน to_exel
with pd.ExcelWriter('New_Data_Excel.xlsx') as writer:
    df6.to_excel(writer, sheet_name='Data_1', index=None)
    df7.to_excel(writer, sheet_name='Data_2', index=None)
```
