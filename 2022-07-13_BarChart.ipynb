```python
import matplotlib.pyplot as plt
import numpy as np


#การวาดกราฟแท่ง
y = [5,15,15,20,10]
x = ['Mon','Tue','Wed','Thu','Fri']

plt.bar(x, y, color='y', alpha=0.8)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.style.use("ggplot")
y = [2,4,1,2,3,2,1,2,3,2]
x = np.arange(1,11)

plt.bar(x, y, color='m')
plt.title("Bar Plot")
plt.xlabel("Play")
plt.ylabel("Score")
plt.grid(color='y', alpha=0.2)
plt.show()
```

```python
# วาดกราฟ Bar Plot แบบข้อมูลหลายชุด

y = [10, 15, 15, 20, 10]
x = np.arange(1,6)

bw = 0.25 #ความกว้างของกราฟแต่ล่พแท่ง

# วาดกราฟแท่งแรกตามปกติ
plt.bar(x, y, color='g', width=bw, label='Product#1')

# กราฟอันที่สอง เลื่อนตำแหน่งในแนวแกน x เท่ากับความกว้างของแท่งกราฟ (อันแรก)
y = [15,25,10,15,20]
plt.bar(x+bw, y, color='lightblue',width=bw, label='Product#2')

# กราฟอันที่สาม เลื่อนตำแหน่งแนวแกน x เท่ากับความกว้างของแท่งกราฟ (อันแรกกับอันที่สองรวมกัน)
y = [10,20,15,10,15]
plt.bar(x+(2*bw), y, hatch='..', color='lightgray', width=bw, label='Product#3')

# เลื่อนระยะ tick ในแนวแกน x เพื่ออยู่ศูนย์กลางของกราฟ
plt.xticks(x+bw, [2016, 2017, 2018, 2019, 2020])

plt.style.use('seaborn')
plt.legend('best')
plt.show()
```
